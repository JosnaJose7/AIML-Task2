```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload the file once; this saves it to /content/
uploaded = files.upload()

# Now you can use the path '/content/Titanic-Dataset.csv' if that is your file name
file_path = '/content/Titanic-Dataset.csv'

df = pd.read_csv(file_path)

def perform_eda(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    summary_stats = df.describe().T
    summary_stats['median'] = df[numeric_cols].median()
    for col in numeric_cols:
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()
    sns.pairplot(df[numeric_cols])
    plt.show()
    corr_matrix = df[numeric_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()
    return summary_stats

summary = perform_eda(df)
print(summary)
```
